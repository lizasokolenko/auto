```python
import os
from nltk import sent_tokenize
import re
import pandas as pd
```

```python
data = open('text.txt')
```

```python
data = data.read()
data = data[:3065]
```

Убираем все ненужные штуки.

```python
data = data.replace('\n', ' ').replace('\xa0', ' ')
```

Убираем лишние пробелы

```python
pat = re.compile(r'\s+')
data1 = pat.sub(' ', data)
data1
```

Разбиваем предложения по регулярке. 

```python
pat1 = re.compile('(?:\.|\!|\?)\s')
data2 = pat1.split(data1)
data2
```

Как мы видим, знаки препинания у нас не сохранились.

Теперь сделаем автоматическое разбиение.

```python
data3 = [re.sub('[\n\t]', ' ', x) for x in sent_tokenize(data1)[:30]]
data3
```

```python
import nltk
nltk.download('punkt')
```

```python
from pprint import pprint
from nltk.tokenize.punkt import PunktSentenceTokenizer, PunktTrainer
```

```python
trainer = PunktTrainer()
trainer.INCLUDE_ALL_COLLOCS = True
trainer.train('\n'.join(data1))
 
tokenizer = PunktSentenceTokenizer(trainer.get_params())
```

Проверяем длину автоматического разбиения текста и разбиения по регуляркам.

```python
print('Tokenizer length - ', len(data3))
print('My split - ', len(data2))
```

Как мы видим, длина массивов одинакова.
Однако эта метрика не очень надежная, т.к. мы не можем узнать, насколько элементы массивов одинаковы по содержанию.
Если мы проверим это метрикой по содержанию, мы получим процент совпадения - 0.

```python
len(set(data2) & set(data3)) / len(set(data2) | set(data3))
```

Сделаем другое автоматическое разбиение.

```python
gold = tokenizer.tokenize(data1)
```

```python
my_split = sent_tokenize(data1)
```

```python
tp = 0
fp = 0
fn = 0

for sent in gold:
    if len(sent_tokenize(sent)) == 1:
        tp += 1
    else:
        fp += 1

for i in range(len(gold)-1):
    sent1, sent2 = gold[i], gold[i+1]
    sent = ' '.join([sent1, sent2])
    if len(sent_tokenize(sent)) == 2:
        tp += 1
    else:
        fn += 1

precision = (tp/(tp+fp))
recall = (tp/(tp+fn))
f1 = 2*(precision*recall)/(precision+recall)
print('Precision - ', precision)
print('Recall - ', recall)
print('F1 - ', f1)
```

Все по единицам, значит массивы совпадают по содержанию.
